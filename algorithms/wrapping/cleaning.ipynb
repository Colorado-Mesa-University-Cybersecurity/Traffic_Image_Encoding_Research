```python
def wrap_samples(df: pd.DataFrame, shape: tuple) -> np.ndarray:
    '''
        Take every sample in the df and convert it into a 2d array with the given shape
         fill in extra space with 0 values
    '''

    # get the values of the dataframe
    vals = df.values

    # pad with columns until we have a column count of the product of the shape tuple
    #    this will allow us to reshape the data into the shape tuple
    if vals.shape[1] < (shape[0] * shape[1]):
        vals = np.pad(vals, ((0,0),(0,shape[0]*shape[1] - vals.shape[1])), 'constant')
    elif vals.shape[1] > (shape[0] * shape[1]):
        print('Error: shape tuple is too small')
        raise ValueError

    # reshape the data into the shape tuple
    vals = vals.reshape((vals.shape[0], shape[0], shape[1]))

    return vals
```

```python
def save_samples_as_images(samples: np.ndarray, path: str, image_names: list) -> None:
    '''
        Save the samples as images
    '''

    # create the path if it doesn't exist
    pathlib.Path(path).mkdir(parents=True, exist_ok=True)

    # iterate through the samples 
    for it,i in enumerate(range(samples.shape[0])):
        # create the image
        img = PIL.Image.fromarray(samples[i], 'L')

        # save the image
        img.save(f'{path}{image_names[it]}_{i}.png')

    return None
```
