# DDoS Dataset Cleaning

```python
import pandas as pd
import os
import numpy as np
```

```python
df = pd.read_csv('/mnt/sda1/CMU_SynTraffic_2022.csv')
```

```python
df['Application Type'].unique()
```

```python
print(df.shape)
```

```python
df.drop(df.loc[df['Data Source']=='CopulaGAN'].index, inplace=True)
df.drop(df.loc[df['Data Source']=='CTGAN'].index, inplace=True)
df.drop(df.loc[df['Data Source']=="VAE"].index, inplace=True)
```

# Rename the Index using tha label names

```python
labels = []
j = 0
for index, row in df.iterrows():
    name = row['Application Type']
    if '_' in name:
        name = name.split('_')[1]
    labels.append(str(j) + "-" + name)
    j += 1
df['Index'] = labels
```

```python
df.drop(['Data Source'],axis=1,inplace=True)
df.drop(['Application Type'],axis=1,inplace=True)
df.drop(['Traffic Type'],axis=1,inplace=True)
```

```python
df['Zero1'] = 0
df['Zero2'] = 0
```

```python
df.to_csv('/mnt/sda1/ddos-data/01-12/cmu-syn-cleaned.csv')
```

# Phase two cleaning

```python
final_df = pd.read_csv('/mnt/sda1/ddos-data/01-12/cmu-syn-cleaned.csv', index_col='Index')
```
